#### Explaination of columns

W - W stands for the win. W column represents the percentage of predicted wins.

R - R stands for the runs. This column represents, the runs scored by player.

AB - AB stands for the At Bat,which means a player is batting.

H - H stands for the Hit. This column represents, the hitting of a ball.

2B - 2B means Double.This column means that after hitting the ball the player safely reached on his base.

3B - 3B stands for the Triple. Triple means that after hitting the ball the player safely reach to third base.

HR - HR stands for the Home Run. After hitting the ball the player should complete all bases and safely reach out to his home.

BB - BB stands for the base on ball.

SO - SO stands for Strikeouts

SB - SB stands for stolen bases

RA - RA stands for Runs Allowed 

ER - ER stands for the Earned Runs which means that the rus earned by a player.

ERA - ERA stands for the Earned Run Average. ERA is an average of run which is earned by a player.

CG - CG stands for the Complete Game.

SHO - SHO stands for the Shutout which is a number of complete games pitched with no runs allowed.

SV - SV stands for the Save. Save is for the player who finishes the game for winning team.

E - E stands for the Errors.

### Removing Outliers

#### Z Score Technique

```python
from scipy.stats import zscore
z = np.abs(zscore(bs))
z.shape
```

```python
bs = bs[(z<3).all(axis=1)]
bs.shape
```

```python
percentage_loss = (30-29)/30*100
percentage_loss
```

```python
x = bs.drop('W',axis=1)
x.shape
```

```python
y = bs['W']
y.shape
```

```python
x.skew().sort_values(ascending=False)
```

### Normalize the data using Log Transformation

```python
x['CG']= np.log(1+x['CG'])
```

```python
x['E'] = np.log(1+x['E'])
x['H'] = np.log(1+x['H'])
```

```python
x.skew().sort_values(ascending=False)
```

```python
x.skew()[np.abs(x.skew()<0.25)].all()
```

### Training Process

```python
from sklearn.preprocessing import MinMaxScaler
mns = MinMaxScaler()

from sklearn.linear_model import LinearRegression,Ridge,Lasso
lr = LinearRegression()

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor

from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_tr = lr.predict(x_train)
    pred_ts = lr.predict(x_test)
    print(f"At random state {i},the training accuracy is:{r2_score(y_train,pred_tr)}")
    print(f"At random state {i},the testing accuracy is:{r2_score(y_test,pred_ts)}")
    print("\n")
```

### Model Testing

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
def train(model,x,y):
    model.fit(x,y)
    pred = model.predict(x)
    cv_score = cross_val_score(model,x,y,scoring = 'neg_mean_squared_error')
    cv_score = np.abs(np.mean(cv_score))
    print('Model Report')
    print('MSE',mean_squared_error(y,pred))
    print('CV',cv_score)
```

```python
model = LinearRegression()
train(model,x,y)
coef = pd.Series(model.coef_,x.columns).sort_values()
coef.plot(kind = 'bar',title = 'Model Coefficients')
```

```python
model = Ridge(normalize=True)
train(model,x,y)
coef = pd.Series(model.coef_,x.columns).sort_values()
coef.plot(kind = 'bar',title = 'Model Coefficients')
```

```python
model = Lasso()
train(model,x,y)
coef = pd.Series(model.coef_,x.columns).sort_values()
coef.plot(kind = 'bar',title = 'Model Coefficients')
```

```python
model = DecisionTreeRegressor()
train(model,x,y)
coef = pd.Series(model.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind = 'bar',title = 'Feature Importance')
```

```python
model = RandomForestRegressor()
train(model,x,y)
coef = pd.Series(model.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind = 'bar',title = 'Feature Importance')
```

```python
model = ExtraTreesRegressor()
train(model,x,y)
coef = pd.Series(model.feature_importances_,x.columns).sort_values(ascending=False)
coef.plot(kind = 'bar',title ='Feature Importance')
```
